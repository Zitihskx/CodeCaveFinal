```python
import os
import pefile
from collections import Counter

def analyze_pe_file(pe_path):
    try:
        pe = pefile.PE(pe_path, fast_load=True)
        
        # Create a list to store section names for the current file
        section_names = []

        # Loop through sections and collect section names
        for section in pe.sections:
            section_names.append(section.Name.decode().rstrip('\x00'))

        return section_names
    except Exception as e:
        print(f"Error processing file {pe_path}: {e}")
        return []

def analyze_directory(directory_path):
    section_counter = Counter()

    # Loop over all files in the directory
    for filename in os.listdir(directory_path):
        file_path = os.path.join(directory_path, filename)
        
        # Only process .exe or .dll files (you can add more file types if necessary)
        if os.path.isfile(file_path):
            print(f"Analyzing file: {file_path}")
            section_names = analyze_pe_file(file_path)
            
            # Update the counter with section names found in this file
            section_counter.update(section_names)

    return section_counter

def print_section_summary(section_counter):
    # Print out the number of unique sections and the number of files containing each section
    print("\nSection Summary:")
    print(f"Total unique section names: {len(section_counter)}\n")
    for section, count in section_counter.items():
        print(f"Section: {section}, Count of files: {count}")

# Specify the directory containing PE files
directory_path = "/home/user/Desktop/Obfus_malware/SelfObfuscate/Petite/TestPetite1/"

# Analyze the directory
section_counter = analyze_directory(directory_path)

# Print the section summary
print_section_summary(section_counter)
```
