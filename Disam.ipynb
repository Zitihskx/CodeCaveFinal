```python
dir_path = '/home/user/Downloads/AttackMalconv2/TrainingData/TestMalware/'
file_list=os.listdir(dir_path)
csv_file_name = '/home/user/Downloads/AttackMalconv2/TrainingData/new_file.csv'
max_size = 0.5*1024*1024
malware_files = []

for root, dirs, files in os.walk(dir_path):
    for file in files:
        file_path = os.path.join(root, file)
        file_size = os.path.getsize(file_path)
        if file_size < max_size:
            malware_files.append(file)
        
```

```python
with open(csv_file_name, 'w', newline='') as csvfile:
    csv_write=csv.writer(csvfile)
    for filess in malware_files:
        csv_write.writerow([filess])
    
```
